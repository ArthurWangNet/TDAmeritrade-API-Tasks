```python
import pandas as pd
import os
import td_config as td
```

```python
folder_1 = '20200730-20200911/'
folder_2 = '20200831-20200923/'
folder_3 = '20200923-20201002/'
```

```python
list_1 = os.listdir(td.UPDATE_MINUTE_DATA_PATH + folder_1)
list_2 = os.listdir(td.UPDATE_MINUTE_DATA_PATH + folder_2)
list_3 = os.listdir(td.UPDATE_MINUTE_DATA_PATH + folder_3)
```

```python
%%timeit
total = len(list_1)
for counter, file in enumerate(list_1):
    print('Working on ' + file+', ' + str(counter) + ' Completed, ' + str(total - counter) + 'Left.')
    df = pd.read_csv(td.UPDATE_MINUTE_DATA_PATH + folder_1 + file)
    df2 = pd.read_csv(td.UPDATE_MINUTE_DATA_PATH + folder_2 + file)
    df3 = pd.read_csv(td.UPDATE_MINUTE_DATA_PATH + folder_3 + file)
    df = df.append(df2)
    df = df.append(df3)
    df.to_csv('./Test/' + file)
print('Completed')
```
